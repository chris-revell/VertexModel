```julia
# Import Julia packages
using DrWatson
@quickactivate
using Revise
using LinearAlgebra
using DelimitedFiles
using SparseArrays
using StaticArrays
using BlockArrays
using CairoMakie
using UnPack
using FromFile
using GeometryBasics
using Random
using Colors
using JLD2
using LaTeXStrings
using Glob
using Printf
using ColorSchemes
using Statistics


@from "$(projectdir())/src/VertexModelContainers.jl" using VertexModelContainers
@from "$(projectdir())/src/Laplacians.jl" using Laplacians
@from "$(projectdir())/src/AnalysisFunctions.jl" using AnalysisFunctions
@from "$(projectdir())/src/CellProperties.jl" using CellProperties
```

```julia

files=Glob.glob("new_energy/100_cells/relaxed/systemData*.jld2","C:\\Users\\v35431nc\\Documents\\VM_code\\VertexModel\\data\\sims" )
saveDir="C:\\Users\\v35431nc\\Documents\\VM_code\\VertexModel\\data\\sims\\new_energy/100_cells/relaxed/"
Γ=zeros(0)
L0=zeros(0)
A_mean=zeros(0)
L_mean=zeros(0)
P_mean=zeros(0)
T_mean=zeros(0)
Peff_mean=zeros(0)
SS_mean=zeros(0)
circ_mean=zeros(0)
for f in files
    @unpack R, matrices, params = load(f)
    @unpack nCells,nVerts, γ, L₀ = params
    @unpack cellAreas,cellPerimeters, cellTensions, cellPressures = matrices


    push!(Γ, mean(γ))
    push!(L0, mean(L₀))
    push!(A_mean,mean(cellAreas))
    push!(L_mean,mean(cellPerimeters))
    push!(P_mean,mean(cellPressures))
    push!(T_mean,mean(-cellTensions))
    shapeParameter=cellPerimeters./sqrt.(cellAreas)
    Peff=getPeff(params, matrices)
    cellQ, cellJ=makeCellQandJ(params, matrices)
    cellShearStress=getShearStress(params, matrices, cellJ)
    cellShapeTensors=makeShapeTensors(R,params,matrices)
    circularity=getCircularity(params, cellShapeTensors)
    push!(Peff_mean,mean(Peff))
    push!(SS_mean,mean(cellShearStress))
    push!(circ_mean,mean(circularity))

    
    
end
```

```julia
fig = Figure()
set_theme!(figure_padding=5, backgroundcolor=(:white,1.0), font="Helvetica", fontsize=19)
ax=Axis(fig[1, 1], ylabel="Γ", xlabel="L0", title="Mean Area")

scatter!(ax,L0,Γ, color=A_mean,markersize=10)

Colorbar(fig[1,2],limits=(minimum(A_mean), maximum(A_mean)),colormap=:viridis,flipaxis=true)

save(datadir(saveDir,"Mean_Area.png"),fig)

fig

fig = Figure()
set_theme!(figure_padding=5, backgroundcolor=(:white,1.0), font="Helvetica", fontsize=19)
ax=Axis(fig[1, 1], ylabel="Γ", xlabel="L0", title="Mean Perimeter")

scatter!(ax,L0,Γ, color=L_mean,markersize=10)

Colorbar(fig[1,2],limits=(minimum(L_mean), maximum(L_mean)),colormap=:viridis,flipaxis=true)

save(datadir(saveDir,"Mean_Perimeter.png"),fig)

fig

fig = Figure()
set_theme!(figure_padding=5, backgroundcolor=(:white,1.0), font="Helvetica", fontsize=19)
ax=Axis(fig[1, 1], ylabel="Γ", xlabel="L0", title="Mean Pressure")

scatter!(ax,L0,Γ, color=P_mean,markersize=10)

Colorbar(fig[1,2],limits=(minimum(P_mean), maximum(P_mean)),colormap=:viridis,flipaxis=true)

save(datadir(saveDir,"Mean_Pressure.png"),fig)

fig

fig = Figure()
set_theme!(figure_padding=5, backgroundcolor=(:white,1.0), font="Helvetica", fontsize=19)
ax=Axis(fig[1, 1], ylabel="Γ", xlabel="L0", title="Mean Tension")

scatter!(ax,L0,Γ, color=T_mean,markersize=10)

Colorbar(fig[1,2],limits=(minimum(T_mean), maximum(T_mean)),colormap=:viridis,flipaxis=true)

save(datadir(saveDir,"Mean_Tension.png"),fig)

fig


fig = Figure()
set_theme!(figure_padding=5, backgroundcolor=(:white,1.0), font="Helvetica", fontsize=19)
ax=Axis(fig[1, 1], ylabel="Γ", xlabel="L0", title="Mean Circularity")

scatter!(ax,L0,Γ, color=circ_mean,markersize=10)

Colorbar(fig[1,2],limits=(minimum(circ_mean), maximum(circ_mean)),colormap=:viridis,flipaxis=true)

save(datadir(saveDir,"Mean_Circularity.png"),fig)

fig
```
