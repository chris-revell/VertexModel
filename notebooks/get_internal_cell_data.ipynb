```julia
f=Glob.glob("Lab_Stuff/Parameter_Inference/Samples/t1_0.01","C:\\Users\\v35431nc\\Documents" )[end]
ff=Glob.glob("a/systemDataFinal*.jld2",f)

for file in ff
    
    @unpack R, matrices, params = load(file)
    @unpack nCells,nVerts, nEdges, γ, L₀ = params

    
    
    F=[1.05 0 ; 0 1/(1.05)]
    Rs=[SVector{2, Float64}(F*R[n]) for n in 1:nVerts]

    spatialData!(Rs,params,matrices)


    @unpack cellAreas,cellPerimeters, cellEdgeCount, cellShapeTensor,A,Ā,B,B̄,C, ϵ = matrices


    edgesP=vec((abs.(ones(nCells)'*B)))
    cellsB=vec(abs.(abs.(B)*edgesP))
    cellsB[findall(x->x!=0,cellsB)].=1
    cellsI=ones(nCells)-cellsB



    evals=eigvals.(cellShapeTensor)
    circ=[abs(x[1]/x[2]) for x in evals]
    shape_param=cellPerimeters./sqrt.(cellAreas)

    df=DataFrame("areas" => cellAreas[cellsI.>0],
                 "perimeters" => cellPerimeters[cellsI.>0], "edge_count" => cellEdgeCount[cellsI.>0], "circularity" => circ[cellsI.>0], "shape_parameter" => shape_param[cellsI.>0])
    df[:, "L_0"] .= L₀
    df[:, "Gamma"] .= γ
    df[:, "sample_set"] .= 'a'


    CSV.write(datadir(f,"stretch_0.05_internal_cells_L0_"*string(L₀)*"_Gamma_"*string(γ)*"_a.csv"), df)

end

```
